# Notebook 18 — RAG Fundamentals
Goal: Understand the basic architecture of Retrieval-Augmented Generation (RAG).

This notebook demonstrates the core pattern:

documents → chunking → embeddings → retrieval → generation

For simplicity we will start with a small set of text documents.

## Section 1 — Install and Import Libraries

```python
# If needed:
# pip install sentence-transformers scikit-learn pandas numpy

import numpy as np
import pandas as pd

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

## Section 2 — Example Documents
In real RAG systems these might come from PDFs, websites, or databases.

```python
documents = [
    "KRAS mutations are common drivers of colorectal cancer.",
    "EGFR inhibitors are used in lung cancer treatment.",
    "BRAF mutations activate the MAPK signaling pathway.",
    "Organoid models are used to test drug response in cancer research.",
    "TP53 is a tumor suppressor gene frequently mutated in cancers."
]

docs_df = pd.DataFrame({"document": documents})
docs_df
```

## Section 3 — Chunking
Large documents are usually split into smaller pieces called chunks.
For this toy example each document is already a small chunk.

```python
chunks = documents
len(chunks)
```

## Section 4 — Create Embeddings
We convert each chunk into a vector embedding using a pretrained model.

```python
model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings = model.encode(chunks)

len(embeddings), embeddings[0].shape
```

## Section 5 — Ask a Query
The user asks a question. We embed the query and find the most relevant chunks.

```python
query = "What gene is commonly mutated in colorectal cancer?"

query_embedding = model.encode([query])
query_embedding.shape
```

## Section 6 — Retrieve Relevant Chunks

```python
similarities = cosine_similarity(query_embedding, embeddings)

similarities
```

```python
top_idx = np.argsort(similarities[0])[::-1][:2]

retrieved_chunks = [chunks[i] for i in top_idx]

retrieved_chunks
```

## Section 7 — Construct a Prompt
In RAG systems the retrieved context is inserted into a prompt for the language model.

```python
context = "\n".join(retrieved_chunks)

prompt = f"""You are a scientific assistant.

Use the context below to answer the question.

Context:
{context}

Question:
{query}

Answer:"""

print(prompt)
```

## Section 8 — Simulated Generation
In a full RAG system the prompt would be sent to an LLM.
For this notebook we simulate the answer using the retrieved context.

```python
answer = "KRAS mutations are common drivers of colorectal cancer."

answer
```

## Section 9 — Evaluate Retrieval
Check whether the retrieved chunks contain the correct information.

```python
for i, chunk in enumerate(retrieved_chunks):
    print(f"Chunk {i+1}: {chunk}")
```
