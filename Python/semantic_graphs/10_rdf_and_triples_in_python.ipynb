# Notebook 10 — RDF and Triples in Python
Goal: Learn the basics of RDF (Resource Description Framework) and represent knowledge graphs as triples using Python and the `rdflib` library.

## Section 1 — Install and Import rdflib

````python
# If needed:
# pip install rdflib

from rdflib import Graph, URIRef, Literal, Namespace
````

## Section 2 — Understanding RDF Triples

RDF represents knowledge as triples:

```
(subject, predicate, object)
```

Example:

```
KRAS participates_in MAPK_pathway
```

## Section 3 — Create an RDF Graph

````python
g = Graph()
len(g)
````

## Section 4 — Define a Namespace

````python
BIO = Namespace("http://example.org/bio/")

BIO
````

## Section 5 — Add Triples

````python
g.add((BIO.KRAS, BIO.participates_in, BIO.MAPK_pathway))
g.add((BIO.Trametinib, BIO.targets, BIO.MEK1))
g.add((BIO.Patient1, BIO.has_mutation, BIO.KRAS))

len(g)
````

### Inspect the graph

````python
for triple in g:
    print(triple)
````

## Section 6 — Add Literal Values

````python
g.add((BIO.KRAS, BIO.symbol, Literal("KRAS")))
g.add((BIO.KRAS, BIO.type, Literal("gene")))

for triple in g:
    print(triple)
````

## Section 7 — Serialize RDF

````python
print(g.serialize(format="turtle"))
````

## Section 8 — Convert NetworkX-style Relationships to RDF

````python
triples = [
    (BIO.KRAS, BIO.participates_in, BIO.MAPK_pathway),
    (BIO.BRAF, BIO.participates_in, BIO.MAPK_pathway),
    (BIO.Cetuximab, BIO.targets, BIO.EGFR)
]

for t in triples:
    g.add(t)

len(g)
````

## Section 9 — Query Triples (Simple Python)

````python
for s,p,o in g:
    if p == BIO.targets:
        print(s, "targets", o)
````
