## Section 1 — Import Libraries

```python
from rdflib import Graph, Namespace, URIRef, Literal
```

## Section 2 — Biomedical Identifier Systems
Real knowledge graphs use **stable identifiers** instead of names.

| Entity | Example |
|------|------|
| Gene | HGNC:6407 |
| Protein | UniProt:P01116 |
| Drug | CHEMBL:CHEMBL120183 |
| Disease | MONDO:0007254 |
| Pathway | Reactome:R-HSA-6802949 |


## Section 3 — Define Namespaces

```python
BIO = Namespace("http://example.org/bio/")
HGNC = Namespace("http://identifiers.org/hgnc/")
CHEMBL = Namespace("http://identifiers.org/chembl/")
REACTOME = Namespace("http://identifiers.org/reactome/")
MONDO = Namespace("http://identifiers.org/mondo/")
```

## Section 4 — Create a Semantic Graph

```python
g = Graph()
```

### Add biological relationships

```python
# Gene participates in pathway
g.add((HGNC["6407"], BIO.participates_in, REACTOME["R-HSA-6802949"]))

# Drug targets gene
g.add((CHEMBL["CHEMBL120183"], BIO.targets, HGNC["3236"]))

# Gene associated with disease
g.add((HGNC["6407"], BIO.associated_with, MONDO["0007254"]))

# Add labels
g.add((HGNC["6407"], BIO.symbol, Literal("KRAS")))
g.add((HGNC["3236"], BIO.symbol, Literal("EGFR")))

len(g)
```

## Section 5 — Inspect the Graph

```python
for triple in g:
    print(triple)
```

## Section 6 — Example SPARQL Query
Which genes are associated with colorectal cancer?

```python
q1 = '''
PREFIX bio: <http://example.org/bio/>
PREFIX mondo: <http://identifiers.org/mondo/>

SELECT ?gene
WHERE {
  ?gene bio:associated_with mondo:0007254 .
}
'''

for row in g.query(q1):
    print(row)
```

## Section 7 — Multi-Hop Semantic Query
Find drugs targeting genes linked to a disease.

```python
q2 = '''
PREFIX bio: <http://example.org/bio/>
PREFIX mondo: <http://identifiers.org/mondo/>

SELECT ?drug
WHERE {
  ?gene bio:associated_with mondo:0007254 .
  ?drug bio:targets ?gene .
}
'''

for row in g.query(q2):
    print(row)
```
