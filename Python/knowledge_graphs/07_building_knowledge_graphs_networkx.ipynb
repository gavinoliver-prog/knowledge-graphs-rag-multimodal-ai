## Section 8 — Build Graph from Table Data

```python
import pandas as pd

data = pd.DataFrame({
    "gene": ["TP53","KRAS","BRAF"],
    "pathway": ["Cell cycle","MAPK pathway","MAPK pathway"]
})

data
```

```python
for _, row in data.iterrows():
    G.add_edge(row["gene"], row["pathway"])

G.edges()
```

## Section 11 — Mini Bio Knowledge Graph

```python
kg = nx.Graph()

kg.add_edge("Patient1","KRAS",relationship="mutation")
kg.add_edge("KRAS","MAPK pathway",relationship="pathway")
kg.add_edge("Trametinib","MEK1",relationship="inhibits")

kg.edges(data=True)
```
