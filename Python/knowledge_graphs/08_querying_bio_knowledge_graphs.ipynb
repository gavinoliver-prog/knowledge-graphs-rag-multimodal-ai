# Notebook 8 — Querying and Exploring Biological Knowledge Graphs
Goal: Explore graph queries, subgraphs, centrality, and simple visualization using NetworkX.

## Section 1 — Setup

```python
import networkx as nx
import matplotlib.pyplot as plt
```

## Section 2 — Create a Small Knowledge Graph

```python
G = nx.Graph()

G.add_edge("Patient1","KRAS",relationship="mutation")
G.add_edge("Patient2","TP53",relationship="mutation")

G.add_edge("KRAS","MAPK pathway",relationship="pathway")
G.add_edge("BRAF","MAPK pathway",relationship="pathway")

G.add_edge("Trametinib","MEK1",relationship="inhibits")
G.add_edge("MEK1","MAPK pathway",relationship="pathway")

G.nodes(), G.edges(data=True)
```

```python
genes = ["KRAS","TP53","BRAF"]
sub = G.subgraph(genes)

sub.nodes(), sub.edges()
```
